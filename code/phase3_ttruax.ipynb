# PHASE 2

## Clustering Using KMeans - No Standardization

#### Create & Fit Model

```python
#Create and fit model
model = KMeans(n_clusters=4).fit(cleansed_data.iloc[:, 1:10])
```

```python
#Calculate cluster label create new dataset
labels = model.labels_
kmeans_data = cleansed_data
kmeans_data['cluster'] = labels
kmeans_data.head()
```

#### Evaluate Cluster Performance

```python
#Crosstabulation
pd.crosstab(kmeans_data['cluster'], kmeans_data['diagnosis'])
```

```python
#Centroids
centroids_array = model.cluster_centers_
centroids_df = pd.DataFrame(centroids_array
                            ,columns = kmeans_data.iloc[:, 1:10].columns
                           )
centroids_df
```

```python
#Inertia
print(model.inertia_)
```

#### Determine Optimal 

```python
#Calculate Inertia
ks = range(1,16)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(kmeans_data.iloc[:, 1:10])
    inertias.append(model.inertia_)
```

```python
#Plot inertia
_ = plt.plot(ks, inertias, '-o', color='#af490d')
_ = plt.title('Inertia by Number of Clusters')
_ = plt.xlabel('number of clusters, k')
_ = plt.ylabel('inertia')
_ = plt.show()
```

Based on the inertia plot, I determine that 3 clusters are optimal.

```python
#Calculate labels & append to dataset
labels = pipeline.predict(kmeans_data.iloc[:, 1:10])
kmeans_data_final = kmeans_data_norm.iloc[:, 0:13]
kmeans_data_final['labels'] = labels
kmeans_data_final.head()
```

#### Evaluate Cluster Performance

```python
#Inertia
pipeline.named_steps['kmeans'].inertia_
```

#### Create Error Rate Function

```python
#Change label values
kmeans_data_final['labels'] = kmeans_data_final['labels'].replace(0, 2)
kmeans_data_final['labels'] = kmeans_data_final['labels'].replace(1, 4)
kmeans_data_final.head(n=15)
```

```python
#Define error rate function
def error_rate(predict_class, actual_class):
    
    error_malig    = sum((predict_class == 4) & (actual_class == 2))
    predict_benign = sum((predict_class == 2))
    b_error_rt     = round((error_malig / predict_benign), 4)
    
    error_benign   = sum((predict_class == 2) & (actual_class == 4))
    predict_malig  = sum((predict_class == 4))
    m_error_rt     = round((error_benign / predict_malig), 4)
    
    error_total    = sum(predict_class != actual_class)
    total_obs      = len(predict_class)
    total_error_rt = round((error_total / total_obs), 4)
    
    print('Error Rate for Benign:', b_error_rt
         ,'\nError Rate for Malignant:', m_error_rt
         ,'\nTotal Error Rate:', total_error_rt)
```

```python
#Call error rate function
error_rate(kmeans_data_final['labels'], kmeans_data_final['class'])
```

#### Report Statement

The error rate of the final clustering in Phase III was relatively low: less than 3% for benign, near 8% for malignant, and around 4.5% overall. However, this approach to generating an error rate is somewhat oversimplified, as it is dependent upon using two and only two clusters.

When comparing the inertia of Phase III to Phase II, we see that three clusters is optimal: Phase III inertia is near 2800, whereas Phase II inertia is around 2300. One way that performance could be compared between Phase II and Phase III would be to create training and testing sets from the data. The performance of the testing sets could be compared to conclude that whichever model performs better on out of sample data is superior.

Additionally, the presence of duplicate values may be introducing bias. Better understanding the source of the initial data, and why these duplicate ids exist would help determine if and how they should be corrected. Furthermore, additional scaling or normalization of the features might help improve performance.
